# Data Preparation and Cleaning



```python
supermarket['Page'].unique()
```

> Page column is drop as it has no attributes in it

```python
sup = supermarket.drop('Page',axis=1)
```

> Drop 'web-scraper-order' column as it is irrelevant

```python
sup.drop('web-scraper-order', axis=1, inplace=True)
```

> Remove "Product Description" from the beginning of each description



```python
# Remove "Product Description" from the beginning of each description
sup['ProductDesc'] = sup['ProductDesc'].str.replace('Product Description', '', regex=False)

# Trim leading and trailing spaces
sup['ProductDesc'] = sup['ProductDesc'].str.strip()

sup.head()
```

```python
sup.info()
```

```python
supermarket['ProductName'].isnull().any()
```

```python
sup[sup['ProductName'].isnull()]
```

> Drop Product Name that has NaN value that will results in irrelevant data as the threshold is more 10 including 'ProductPrice', 'ProductSold', 'ProductRating', 'ProductDesc', 'ProductImg1-src', 'Product_5star', 'Product_4Star', 'Product_3Star', Product_2Star', 'Product_1Star' columns.

```python
sup.dropna(axis=0, subset=['ProductName'], inplace=True)
```

```python
sup['ProductName'].isnull().any()
```

```python
sup['ProductRating'].isnull().any()
```

```python
sup[sup['ProductRating'].isnull()]
```

> Extract the Product Star Rating to include only numeric.

```python
# List of columns to extract and multiply
columns_to_process = ['Product_5star', 'Product_4Star', 'Product_3Star', 'Product_2Star', 'Product_1Star']

# Set the display format for float values
pd.options.display.float_format = '{:.0f}'.format

# Iterate over the columns and extract, trim, and convert values
for col in columns_to_process:
    extracted_values = sup[col].str.extract(r'\(([\d.]+)([k]*)\)', expand=False)
    numeric_values = extracted_values[0].astype(float)  # Convert extracted values to float
    k_indicator = extracted_values[1].apply(lambda x: 1000 if x == 'k' else 1)  # Get multiplier for 'k'
    sup[col] = numeric_values * k_indicator

# Checking the updated DataFrame
sup
```

> Check if the 17k in 'Product_5star' column has conduct valid calculation.

> eg. 17k become 17000

```python
desired_value = 1700

row = sup[sup['Product_5star'] == desired_value]

row
```

> As 'ProductRating' column has NaN because of it has all '0' for 'Product_5star', 'Product_4Star', 'Product_3Star', 'Product_2Star', 'Product_1Star', thus we will fill the NaN values with 0

```python
sup['ProductRating'].fillna(0, inplace=True)

sup
```

> Check if all the ProductRating = 0 has '0' for all stars

```python
desired_value = 0

row = sup[sup['ProductRating'] == desired_value]

row
```

```python
sup['ProductSold'].info()
```

> Convert k in ProductSold to thousands.

> eg. 4.4k to 4400

```python
# Convert the "ProductSold" column to string type
sup['ProductSold'] = sup['ProductSold'].astype(str)

# Multiply values by 1000 if they contain 'k'
sup['ProductSold'] = sup['ProductSold'].apply(lambda x: int(float(x.replace('k', '')) * 1000) if 'k' in x else int(x))

# Convert the "ProductSold" column to integer type
sup['ProductSold'] = sup['ProductSold'].astype(int)

# Print the updated DataFrame
sup
```

- Then, we replace 0 in **Pekerjaan** column to **TIADA** to standardize the data.


```python
sup['ProductSold'].unique()
```

> In ProductPrice column, remove RM.

> If there is range of price, take the maximum price and remove the RM.

```python
sup['ProductPrice'] = sup['ProductPrice'].str.replace('RM', '')
sup['ProductPrice'] = sup['ProductPrice'].str.split(' - ').apply(lambda x: max([float(val) for val in x]) if isinstance(x, list) else float(x))
sup['ProductPrice'] = sup['ProductPrice'].round(2)
sup['ProductPrice'] = sup['ProductPrice'].astype(str)
sup['ProductPrice'] = sup['ProductPrice'].str.replace('.0', '')
sup['ProductPrice'] = sup['ProductPrice'].str.replace('.', '.')
sup['ProductPrice'] = sup['ProductPrice'].astype(str)
```
