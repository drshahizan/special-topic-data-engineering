Fetch Fresh Apples Image

```python
def fetch_images_with_pagination():
    folder_name = "Fresh Apples"
    max_results_per_page = 50
    next_cursor = None
    all_images = []

    while True:
        # Make request with pagination parameters
        params = {
            'max_results': max_results_per_page,
            'next_cursor': next_cursor
        }
        response = cloudinary.api.resources(**params)

        # Process the response
        images = response['resources']

        # Filter images by folder_name
        folder_images = [image for image in images if image['folder'] == folder_name]
        all_images.extend(folder_images)

        # Check if there is a next page
        if 'next_cursor' in response:
            next_cursor = response['next_cursor']
        else:
            break

    return all_images

# Usage
cloudinary.config(
    cloud_name='dbozk6mei',
    api_key='489422932326316',
    api_secret='vcTE2WuFoWYM8Uo9biFV82VhtM0')

images = fetch_images_with_pagination()

# Create a new folder called "Fresh Apple" in the current directory
folder_name = "Fresh Apple"
os.makedirs(folder_name, exist_ok=True)

# Save images to the "Fresh Apple" folder
for i, image in enumerate(images):
    url = image['url']
    file_name = f"{folder_name}/image_{i}.png"
    urllib.request.urlretrieve(url, file_name)
    print(f"Image {i+1}/{len(images)} saved as {file_name}")

print(f"Total images fetched: {len(images)}")
```

# Data Preprocessing & Cleaning

Resizing

```python
from PIL import Image
import os

# Define the target size for resizing the images
target_size = (224, 224)  # Adjust as per your requirements

# Function to resize images
def resize_images(directory):
    for filename in os.listdir(directory):
        file_path = os.path.join(directory, filename)
        if os.path.isfile(file_path):
            try:
                # Open the image file
                image = Image.open(file_path)

                # Resize the image
                resized_image = image.resize(target_size)

                # Save the resized image, overwriting the original file
                resized_image.save(file_path)
            except Exception as e:
                print(f"Error resizing image {filename}: {e}")

# Provide the path to each folder containing the images
folders = ["Fresh Apple", "Fresh Orange", "Fresh Banana",
           "Rotten Apple", "Rotten Orange", "Rotten Banana"]

# Resize images in each folder
for folder in folders:
    resize_images(folder)
```

Split to Test, Train and Validation folders

```python
import os
import random
import shutil

# Set the random seed for reproducibility
random.seed(42)

# Set the path to the parent directory containing the class folders
parent_dir = "/content"

# Set the train-test-validation split ratios
train_ratio = 0.6
test_ratio = 0.2
val_ratio = 0.2

# Create train, test, and validation folders
train_dir = os.path.join(parent_dir, "train")
test_dir = os.path.join(parent_dir, "test")
val_dir = os.path.join(parent_dir, "validation")
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)

# List the class folders
class_folders = ["Fresh Apple", "Fresh Orange", "Fresh Banana",
                 "Rotten Apple", "Rotten Orange", "Rotten Banana"]

for folder in class_folders:
    # Set the path to the class folder
    class_folder = os.path.join(parent_dir, folder)

    # Create train, test, and validation subfolders for the class
    train_class_folder = os.path.join(train_dir, folder)
    test_class_folder = os.path.join(test_dir, folder)
    val_class_folder = os.path.join(val_dir, folder)
    os.makedirs(train_class_folder, exist_ok=True)
    os.makedirs(test_class_folder, exist_ok=True)
    os.makedirs(val_class_folder, exist_ok=True)

    # List all images in the class folder
    images = os.listdir(class_folder)

    # Shuffle the images randomly
    random.shuffle(images)

    # Split the images into train, test, and validation sets
    num_images = len(images)
    num_train_images = int(num_images * train_ratio)
    num_test_images = int(num_images * test_ratio)
    train_images = images[:num_train_images]
    test_images = images[num_train_images:num_train_images + num_test_images]
    val_images = images[num_train_images + num_test_images:]

    # Move the train images to the train class folder
    for image in train_images:
        src_path = os.path.join(class_folder, image)
        dest_path = os.path.join(train_class_folder, image)
        shutil.copy(src_path, dest_path)

    # Move the test images to the test class folder
    for image in test_images:
        src_path = os.path.join(class_folder, image)
        dest_path = os.path.join(test_class_folder, image)
        shutil.copy(src_path, dest_path)

    # Move the validation images to the validation class folder
    for image in val_images:
        src_path = os.path.join(class_folder, image)
        dest_path = os.path.join(val_class_folder, image)
        shutil.copy(src_path, dest_path)
```

# Data Analysis & Visualization (EDA)

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the folder names for different categories
folders = ["Fresh Apple", "Fresh Orange", "Fresh Banana",
           "Rotten Apple", "Rotten Orange", "Rotten Banana"]

# Create an empty DataFrame to store the image details
data = pd.DataFrame(columns=["Category", "Image Path"])

# Iterate over each folder and gather image details
for folder in folders:
    category = folder.split()[-1]  # Extract the category name
    image_paths = [os.path.join(folder, filename) for filename in os.listdir(folder)]

    # Append the category and image paths to the DataFrame
    category_data = pd.DataFrame({"Category": category, "Image Path": image_paths})
    data = data.append(category_data, ignore_index=True)

# Print the first few rows of the DataFrame
print("Image Data:")
print(data.head())

# Calculate the count of images in each category
category_counts = data["Category"].value_counts()

# Visualize the distribution of categories
plt.figure(figsize=(8, 6))
sns.countplot(x="Category", data=data)
plt.title("Distribution of Fruit Categories")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()

# Print the count of images in each category
print("\nCategory Counts:")
print(category_counts)
```

```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the folder names for different categories in the train, test, and validation folders
train_folders = ["train/Fresh Apple", "train/Fresh Orange", "train/Fresh Banana",
                 "train/Rotten Apple", "train/Rotten Orange", "train/Rotten Banana"]

test_folders = ["test/Fresh Apple", "test/Fresh Orange", "test/Fresh Banana",
                "test/Rotten Apple", "test/Rotten Orange", "test/Rotten Banana"]

validation_folders = ["validation/Fresh Apple", "validation/Fresh Orange", "validation/Fresh Banana",
                      "validation/Rotten Apple", "validation/Rotten Orange", "validation/Rotten Banana"]

# Create an empty DataFrame to store the image details
data_train = pd.DataFrame(columns=["Category", "Image Path"])
data_test = pd.DataFrame(columns=["Category", "Image Path"])
data_validation = pd.DataFrame(columns=["Category", "Image Path"])

# Iterate over each folder in the train folder and gather image details
for folder in train_folders:
    category = folder.split("/")[-1]  # Extract the category name
    image_paths = glob.glob(f"{folder}/*.png")  # Assumes all images are in PNG format

    # Append the category and image paths to the train DataFrame
    category_data = pd.DataFrame({"Category": category, "Image Path": image_paths})
    data_train = pd.concat([data_train, category_data], ignore_index=True)

# Iterate over each folder in the test folder and gather image details
for folder in test_folders:
    category = folder.split("/")[-1]  # Extract the category name
    image_paths = glob.glob(f"{folder}/*.png")  # Assumes all images are in PNG format

    # Append the category and image paths to the test DataFrame
    category_data = pd.DataFrame({"Category": category, "Image Path": image_paths})
    data_test = pd.concat([data_test, category_data], ignore_index=True)

# Iterate over each folder in the validation folder and gather image details
for folder in validation_folders:
    category = folder.split("/")[-1]  # Extract the category name
    image_paths = glob.glob(f"{folder}/*.png")  # Assumes all images are in PNG format

    # Append the category and image paths to the validation DataFrame
    category_data = pd.DataFrame({"Category": category, "Image Path": image_paths})
    data_validation = pd.concat([data_validation, category_data], ignore_index=True)

# Print the first few rows of the train DataFrame
print("Train Image Data:")
print(data_train.head())

# Print the first few rows of the test DataFrame
print("Test Image Data:")
print(data_test.head())

# Print the first few rows of the validation DataFrame
print("Validation Image Data:")
print(data_validation.head())

# Calculate the count of images in each category in the train DataFrame
train_category_counts = data_train["Category"].value_counts()

# Calculate the count of images in each category in the test DataFrame
test_category_counts = data_test["Category"].value_counts()

# Calculate the count of images in each category in the validation DataFrame
validation_category_counts = data_validation["Category"].value_counts()

# Visualize the distribution of categories in the train DataFrame
plt.figure(figsize=(8, 6))
sns.countplot(x="Category", data=data_train)
plt.title("Distribution of Fruit Categories in Train Data")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()

# Visualize the distribution of categories in the test DataFrame
plt.figure(figsize=(8, 6))
sns.countplot(x="Category", data=data_test)
plt.title("Distribution of Fruit Categories in Test Data")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()

# Visualize the distribution of categories in the validation DataFrame
plt.figure(figsize=(8, 6))
sns.countplot(x="Category", data=data_validation)
plt.title("Distribution of Fruit Categories in Validation Data")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()

# Print the count of images in each category in the train DataFrame
print("\nTrain Category Counts:")
print(train_category_counts)

# Print the count of images in each category in the test DataFrame
print("\nTest Category Counts:")
print(test_category_counts)

# Print the count of images in each category in the validation DataFrame
print("\nValidation Category Counts:")
print(validation_category_counts)
```

# Model Training

## Convolutional Neural Network



```python
import os
import random
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import VGG16
from sklearn.metrics import classification_report, confusion_matrix

# Set the random seed for reproducibility
random.seed(42)
np.random.seed(42)
tf.random.set_seed(42)

# Set the paths to the train and test folders
train_dir = "/content/train"
test_dir = "/content/test"

# Set the number of classes and batch size
num_classes = 6
batch_size = 32

# Preprocess and augment the data
train_datagen = ImageDataGenerator(rescale=1./255,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(train_dir,
                                                    target_size=(224, 224),
                                                    batch_size=batch_size,
                                                    class_mode='categorical')

test_generator = test_datagen.flow_from_directory(test_dir,
                                                  target_size=(224, 224),
                                                  batch_size=batch_size,
                                                  class_mode='categorical')

# Load the pre-trained VGG16 model without the top (fully connected) layers
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze the base model's layers
for layer in base_model.layers:
    layer.trainable = False

# Create the top layers for the classification task
model = tf.keras.Sequential([
    base_model,
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(num_classes, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
model.fit(train_generator,
          steps_per_epoch=train_generator.samples // batch_size,
          epochs=20)

# Train the model and track the accuracy over epochs
history = model.fit(train_generator,
                    steps_per_epoch=train_generator.samples // batch_size,
                    epochs=5,
                    validation_data=test_generator,
                    validation_steps=test_generator.samples // batch_size)
```

```python
# Plot accuracy over epochs
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
# Plot loss over epochs
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```
