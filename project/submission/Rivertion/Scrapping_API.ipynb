```python
import cloudinary
import cloudinary.uploader
import requests
import pymongo
import os

API_KEY = "35339771-b462e4b277b61d82176e7e5b8"
SEARCH_TERM = "banana"
RESULTS_PER_PAGE = 100

# Creating a folder to save the images and assigning to a variable for further use
Image_Folder = 'Rivertion'
Cloudinary_Folder = 'Rivertion'

def main():
    if not os.path.exists(Image_Folder):
        os.mkdir(Image_Folder)
    download_images()

# Connect to the MongoDB server
client = pymongo.MongoClient("mongodb+srv://mincridible:minzpro1@min.tan7fdn.mongodb.net/?retryWrites=true&w=majority")
db = client['Rivertion']
collection = db['Fruits']

# Configure Cloudinary
cloudinary.config(
  cloud_name = "dysm5zxli",
  api_key = "866311266618576",
  api_secret = "rOhbO2InQs_RLoDpQa9898mtYOg"
)


def download_images():
    url = f"https://pixabay.com/api/?key={API_KEY}&q={SEARCH_TERM}&per_page={RESULTS_PER_PAGE}"
    response = requests.get(url)
    json_data = response.json()

    for image in json_data['hits']:
        name = image['id']
        img_url = image["webformatURL"]
        r = requests.get(img_url, stream=True)
        with open(Image_Folder + '/' + str(name) + '.jpg', 'wb') as f:
            f.write(r.content)

        # Upload the image to Cloudinary with the specified folder
        upload_result = cloudinary.uploader.upload(
            Image_Folder + '/' + str(name) + '.jpg',
            folder=Cloudinary_Folder
        )

        # Create a document with metadata and Cloudinary details
        doc = {
            'filename': str(name) + '.jpg',
            'id': image['id'],
            'imageWidth': image['imageWidth'],
            'imageHeight': image['imageHeight'],
            'previewWidth':image['previewWidth'],
            'previewHeight':image['previewHeight'],
            'webformatWidth':image['webformatWidth'],
            'webformatHeight':image['webformatHeight'],
            'imageSize': image['imageSize'],
            'type' : image['type'],
            'tags' : image['tags'],
            'view': image['views'],
            'downloads': image['downloads'],
            'likes':image['likes'],
            'comments': image['comments'],
            'user_id':image['user_id'],
            'user':image['user'],
            'pageURL': image['pageURL'],
            'previewURL': image['previewURL'],
            'userImageURL': image['userImageURL'],
            'webformatURL': image['webformatURL'],
            'cloudinary_public_id': upload_result['public_id'],
            'cloudinary_url': upload_result['secure_url']
        }

        # Insert the document into the collection
        collection.insert_one(doc)

        document_count = collection.count_documents({})
        print(f"Number of documents in 'Fruits' collection: {document_count}")

if __name__ == '__main__':
    main()
```
